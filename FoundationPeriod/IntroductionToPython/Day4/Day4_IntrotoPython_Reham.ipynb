### Using error handling, write a fuction that takes in a string and retruns the summation of digits in the string. 
### Example: ''a1b2c3'' should return 6 

```python
def str_count(s):
    summation=0
    for i in s :
        try:
            summation+=(int(i))
        except:
            pass
    print(summation)
s='a1b2c3'
str_count(s)
```

## Write a program that reads the file ''sampleText.txt'' and couts the number of time letter T/t occured in the file 

```python
fil=open("sampleText.txt")
counter=0
for i in fil:
    for j in i:
        if j=='T' or j=='t':
            counter+=1
print(counter)
```

## Reading and Writing json files

```python
#json.dumps() converts a Python dictionary to a json string
import json
dic= {'id':[1,2,3],'name':['Ahmed','Mohamed','Omar']}
file = open('output.json','w' )
json_string=json.dumps(dic)
file.write(json_string)
print(json.dumps(dic))
file.close()
```

```python
#json.loads() converts a json string to a dictionary
output =json.loads(json_string)
print(output)
print(type(output))
```

## Write a function that prints all the values associated with the key name in sampleJson

```python
import json
def func(key,diction):
    for i in diction:
        print(i[key])
        
sampleJson = """[
   { 
      "id":1,
      "name":"name1",
      "color":[ 
         "red",
         "green"
      ]
   },
   { 
      "id":2,
      "name":"name2",
      "color":[ 
         "pink",
         "yellow"
      ]}
      ]"""


output =json.loads(sampleJson)
func("name",output)
```

## Get the average of a numerical column

```python
import csv
data = csv.DictReader(open("cardioactivities.csv")) # convert csv file to dictionary
ls=[]
for i in data:
    #print(i["Average Heart Rate (bpm)"])
    ls.append(float(i["Average Heart Rate (bpm)"]))
#print(ls)
print(sum(ls)/len(ls))
```
